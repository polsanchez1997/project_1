```python
import pandas as pd 

data = pd.read_csv("data/attacks.csv", encoding = 'ISO-8859-1', dtype=object)
data
```

Cleaning Data

First of all, we need to make sure if there are rows without value and, if so, we need to delete them.

```python
shark = data.dropna(how="all")
shark
```

After cleaning thousands of rows, we need to make sure that there are no duplicates. That is why we proceed to avoid duplicates:

```python
shark_= shark.drop_duplicates()
shark_
```

We managed to clean more than 2000 duplicates. However, it is also appreciated that some of them have '0' and 'xx' as their case number value. We proceed to clean it by deleting those rows:

```python

shark_ = shark_.drop(shark_[shark_['Case Number'] == '0'].index)
shark_ = shark_.drop(shark_[shark_['Case Number'] == 'xx'].index)
shark_
```

After this, we can see that there are two columns that could be noise: "Unnamed: 22" and "Unnamed:23". We will check is this is true, and if so, we will remove them.

```python
shark_["Unnamed: 22"].value_counts()
shark_["Unnamed: 23"].value_counts()
```

```python
new_shark = shark_.drop(["Unnamed: 22", "Unnamed: 23"], axis=1)
new_shark
```
