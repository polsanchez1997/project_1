```python
import pandas as pd 
import numpy as np
data = pd.read_csv("data/attacks.csv", encoding = 'ISO-8859-1', dtype=object)
data
```

### Cleaning Data

First of all, we need to make sure if there are rows without value and, if so, we need to delete them.

```python
shark = data.dropna(how="all")
shark
```

After cleaning thousands of rows, we need to make sure that there are no duplicates. That is why we proceed to avoid duplicates:

```python
shark_= shark.drop_duplicates()
shark_
```

We managed to clean more than 2000 duplicates. However, it is also appreciated that some of them have '0' and 'xx' as their case number value. We proceed to clean it by deleting those rows:

```python

shark_ = shark_.drop(shark_[shark_['Case Number'] == '0'].index)
shark_ = shark_.drop(shark_[shark_['Case Number'] == 'xx'].index)
shark_
```

After this, we can see that there are two columns that could be noise: "Unnamed: 22" and "Unnamed:23". We will check is this is true, and if so, we will remove them.

```python
shark_["Unnamed: 22"].value_counts()
shark_["Unnamed: 23"].value_counts()
```

```python
new_shark = shark_.drop(["Unnamed: 22", "Unnamed: 23"], axis=1)
new_shark
```

It is important to rename the columns so we work with standarized data:

```python
new_shark.dict = {i : i.lower().replace (" ", "_") for i in new_shark.columns}
new_shark.rename (columns = new_shark.dict, inplace = True)
new_shark
```

Now, we are going to work with the "sex" column as we will need that one for visualization. We will first rename and then work on it creating a function:

```python
new_shark.rename(columns={'sex_': 'sex'}, inplace=True)
```

```python

def cleaning_sex(value):
    
    if type(value) == str and (value == 'M' or value == 'F'):
        return value.strip()
    return 'unknown'

def sex_cleaning():
    new_shark['sex'] = new_shark['sex'].map(cleaning_sex)

sex_cleaning()
```

```python
new_shark["sex"].value_counts()
```

We will proceed the same way with the "fatal" column.

```python
new_shark.rename(columns={"fatal_(y/n)": "fatal"}, inplace=True)
new_shark["fatal"] = new_shark["fatal"].str.upper()

def cleaning_fatal(value):
    if value == 'N':
        return 'N'
    elif value == 'Y':
        return 'Y'
    else:
        return 'unknown'

def fatal_cleaning():
    new_shark["fatal"] = new_shark["fatal"].map(cleaning_fatal)


fatal_cleaning()
```

Time to clean the activity column. This one is tricky since every row has different values, so we need to create categories.

```python
def activity_standarize(str_):
    
    if pd.isna(str_):
        return "others"

    str_lower = str_.lower()
    if 'swim' in str_lower:
        return "swimming"
    if any(activity in str_lower for activity in ['bath', 'float', 'crouch', 'dangl',]):
        return "bathing"
    if any(activity in str_lower for activity in ['fish', 'hunt', 'net']):
        return "fishing"
    if any(activity in str_lower for activity in ['natural']):
        return "natural disasters"
    if any(activity in str_lower for activity in ['surf', 'board', 'paddel', 'tread']):
        return "surfing"
    if any(activity in str_lower for activity in ['air', 'parachut']):
        return "air accident"
    if any(activity in str_lower for activity in ['boat', 'ship', 'wreck', 'yacht', 'submarine','sink']):
        return "sailing"
    if 'div' in str_lower:
        return "diving"

    return "others"

new_shark["activity"] = new_shark["activity"].apply(lambda x: activity_standarize (x))
new_shark["activity"].value_counts()
```

```python
death_by_activity = new_shark[(new_shark["activity"].notna()) & (new_shark["fatal"] == "Y")]
death_by_activity = death_by_activity[["activity",]]
death_by_activity
```

Lastly, we need to do the same with time. I decided to work with integers as it was easier.

```python
def time_categorize (x):
    if isinstance(x,str):
        pos_h = x.find("h") 
        if pos_h > 0: 
            time_extraction = x[:pos_h]
            if time_extraction.isdigit(): 
                time_int = int(time_extraction)
                if 4 <= time_int < 12:
                    return ("Morning")
                elif 12 <= time_int < 20:
                    return ("Afternoon")
                else:
                    return ("Night")    
            else: 
                return("Unknown")
        else:
            return("Unknown")
    return("Unknown")

new_shark["time"] = new_shark["time"].apply(lambda x: time_categorize (x))
```

```python
new_shark.to_csv('new_shark.csv', index=False)
```
